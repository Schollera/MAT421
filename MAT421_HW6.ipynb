**Continuity**:

Continuity is the mathematical property where a function, f(x), does not exhibit abrupt jumps, breaks, or discontinuities. We test continuity as a point x=a if the function is defined at a, and the function's value at a equals the limit value as x approaches a (from both sides of the limit).



```python
import numpy as np
import matplotlib.pyplot as plt

# Define function
def f(x):
    return np.sin(x) / x if x != 0 else 1  # Singularity at x = 0

# Define limit
def limit_of_f(x_value):
    epsilon = 1e-5
    return (f(x_value + epsilon) + f(x_value - epsilon)) / 2

# Plot function and the limit
x_values = np.linspace(-10, 10, 400)
y_values = [f(x) for x in x_values]
plt.plot(x_values, y_values, label='f(x) = sin(x)/x')

# Check limit of 0
x_value = 0
plt.axvline(x=x_value, color='r', linestyle='--', label=f'Limit at x = {x_value}')
plt.legend()
plt.title('Continuity of the function f(x) = sin(x)/x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()

# Calculate limit at 0
print(f"Limit of f(x) at x = 0: {limit_of_f(0)}")
```

**Differentiation**:

Differentiation is the concept that deals with finding the rate of change of a function, f(x). The derivative at point x=a represents how the function's output changes as the input changes at or around a.

```python
import sympy as sp

# Define variable
x = sp.symbols('x')

# Define function
f = sp.sin(x) * sp.exp(x)

# Compute derivative
f_prime = sp.diff(f, x)

# Display the derivative
print(f"The derivative of f(x) = sin(x) * exp(x) is: {f_prime}")

# Plot function and derivative
import numpy as np
import matplotlib.pyplot as plt

f_lambdified = sp.lambdify(x, f, "numpy")
f_prime_lambdified = sp.lambdify(x, f_prime, "numpy")

x_vals = np.linspace(-10, 10, 400)
y_vals = f_lambdified(x_vals)
y_prime_vals = f_prime_lambdified(x_vals)

plt.plot(x_vals, y_vals, label='f(x) = sin(x) * exp(x)')
plt.plot(x_vals, y_prime_vals, label="f'(x) = Derivative of f(x)", linestyle='--')
plt.legend()
plt.title('Function and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**Taylor's Theorem**:

Taylor's Theorem is a method to approximate a function, f(x), through a polynomial representation around a point, x=a. The function is depicted as an infinite sum derived from its derivatives. The expansion around a=0 is noted as the Maclaurian series.

```python
import sympy as sp

# Define variable and function
x = sp.symbols('x')
f = sp.exp(x)

# Compute Taylor expansion of exp(x) around x=0 to the 4th degree
taylor_expansion = sp.series(f, x, 0, 5).removeO()

print(f"Taylor series expansion of exp(x) around x = 0: {taylor_expansion}")

# Plot function and the expansion
import numpy as np
import matplotlib.pyplot as plt

f_lambdified = sp.lambdify(x, f, "numpy")
taylor_expansion_lambdified = sp.lambdify(x, taylor_expansion, "numpy")

x_vals = np.linspace(-2, 2, 400)
y_vals = f_lambdified(x_vals)
y_taylor_vals = taylor_expansion_lambdified(x_vals)

plt.plot(x_vals, y_vals, label='exp(x)')
plt.plot(x_vals, y_taylor_vals, label="Taylor Series Approximation", linestyle='--')
plt.legend()
plt.title('Function and Taylor Expansion (degree 4)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**Optimization with Gradient Descent**:

Optimization with gradient descent is the algorithm of finding a maximum/minimum of a function, f(x). We leverage the step size to control how far the iteration moves towards the local minimum/maxmimum value. The gradient is leveraged, and we move from the initial guess to the convergance value.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function as f(x) = (x - 3)^2
def f(x):
    return (x - 3)**2

# Define gradient
def grad_f(x):
    return 2 * (x - 3)

# Optimization Algorithm
def gradient_descent(learning_rate=0.1, n_iterations=100, initial_point=0):
    x = initial_point
    history = [x]
    for _ in range(n_iterations):
        x = x - learning_rate * grad_f(x)
        history.append(x)
    return history

# Run Algorithm
history = gradient_descent(learning_rate=0.1, n_iterations=50)

# Plot function and gradient descent
x_vals = np.linspace(-2, 8, 40)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label='f(x) = (x - 3)^2')
plt.scatter(history, f(np.array(history)), color='r', label='Gradient Descent Steps')
plt.legend()
plt.title('Optimization with Gradient Descent')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
