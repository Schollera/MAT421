Linear Interpolation is a method of estimating unknown values between two known values simply with a straight line.

```python
import numpy as np
import matplotlib.pyplot as plt

#Data
x = np.array([0, 1, 3, 4])
y = np.array([0, 1, 9, 16])

#Linear interpolation
from scipy.interpolate import interp1d
linear_interpolation = interp1d(x, y, kind='linear')

#finer x range for plotting
x_finer = np.linspace(0, 4, 100)
y_finer = linear_interpolation(x_finer)

#Plot
plt.plot(x, y, 'o', label="Data points")
plt.plot(x_finer, y_finer, '-', label="Linear Interpolation")
plt.legend()
plt.title('Linear Interpolation')
plt.show()
```

Cubic spline interpolation is similar to linear interpolation. Yet, it utilizes a piecewise cubic polynomial to interpolate the data, leading to smoother curves in-between points.

```python
from scipy.interpolate import CubicSpline

#data
x = np.array([0, 1, 3, 4])
y = np.array([0, 1, 9, 16])

# Perform cubic spline interpolation
cubic_spline_interpolation = CubicSpline(x, y)

# Create a finer x range for smooth plotting
x_finer = np.linspace(0, 4, 100)
y_finer = cubic_spline_interpolation(x_finer)

# Plot
plt.plot(x, y, 'o', label="Data points")
plt.plot(x_finer, y_finer, '-', label="Cubic Spline Interpolation")
plt.legend()
plt.title('Cubic Spline Interpolation')
plt.show()
```

General interpolation may also be utilized in the scipy package with the interp1d function. This may create a linear, quadratic, or cubic interpolation.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

#data
x = np.array([0, 1, 3, 4])
y = np.array([0, 1, 9, 20])

# General interpolation -- quadratic form
general_interp = interp1d(x, y, kind='quadratic')

#finer x range for plotting
x_finer = np.linspace(0, 4, 100)
y_finer = general_interp(x_finer)

#Plot
plt.plot(x, y, 'o', label="Data points")
plt.plot(x_finer, y_finer, '-', label="General Interpolation -- Quadratic")
plt.legend()
plt.title('General Interpolation -- Quadratic')
plt.show()
```

Lagrange polynomial interpolation constructs a polynomial with the lagrange technique. Yet, it may be prone to overfitting.

```python
from scipy.interpolate import lagrange

#data
x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.array([0, 1, 4, 9, 24, 25, 40])

#Lagrange interpolation
lagrange_poly = lagrange(x, y)

#plot x range more finely
x_finer = np.linspace(0, 6, 500)
y_finer = lagrange_poly(x_finer)

#Plot
plt.plot(x, y, 'o', label="Data points")
plt.plot(x_finer, y_finer, '-', label="Lagrange Polynomial Interpolation")
plt.legend()
plt.title('Lagrange Polynomial Interpolation')
plt.show()
```
