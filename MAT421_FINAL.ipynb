**Literature Review**

*Definitions and Background*

Unemployment is a critical macroeconomic indicator defined as the proportion of individuals in the labor force actively seeking but unable to find employment. It directly reflects economic health, signaling resource utilization, economic stability, and societal welfare (Zhang, 2023; Reid, 2023). Various macroeconomic variables influence unemployment, including real GDP, the Consumer Price Index (CPI), interest rates, and inflation, each contributing uniquely to labor market dynamics.

Real Gross Domestic Product (GDP) is the total market value of all final goods and services produced within an economy, adjusted for inflation. It measures economic output and growth, correlating negatively with unemployment due to demand-side effects. Economic expansions (higher real GDP) typically correlate with lower unemployment rates, as increased production necessitates more labor (Green et al., 2023).

The Consumer Price Index (CPI) quantifies the average change over time in the prices consumers pay for a basket of goods and services, thus directly indicating inflation levels. Inflation, a sustained increase in overall price levels, has intricate relationships with unemployment, often conceptualized through the Phillips Curve. The Phillips Curve posits a trade-off between unemployment and inflation—typically, as unemployment decreases, inflation increases due to wage pressures and increased consumer demand (Smith, Timmermann, & Wright, 2024).

Interest rates represent the cost of borrowing money, set primarily by central banks to control monetary policy. Interest rates indirectly influence employment by affecting investment levels, consumer spending, and aggregate demand. Typically, higher interest rates slow economic growth and increase unemployment, whereas lower rates encourage economic expansion and employment growth (Ahmed & Nasser, 2023; Reid, 2023). Understanding their dynamics is essential for policymakers to implement effective stabilization strategies.

*Current Empirical Evidence*

Green et al. (2023) provide an extensive empirical analysis exploring macroeconomic and microeconomic determinants of U.S. unemployment from 2000 onwards. Employing linear regression, the authors identify GDP (negative effect), inflation (positive effect), unemployment insurance (negative, contrary to common belief), and the S&P 500 index (negative effect) as significant macroeconomic predictors. The study uniquely integrates microeconomic variables, demonstrating that higher education and health disparities significantly impact individual unemployment likelihood, with pronounced racial disparities existing despite other controls. Methodologically, this comprehensive regression approach allows the exploration of macroeconomic and socio-demographic determinants.

Reid (2023) conducts a robust macroeconomic analysis using quarterly U.S. data from 1948 to 2021 to estimate determinants of aggregate employment. The study leverages multiple regression techniques, focusing primarily on expected demand proxies—personal consumption expenditures (PCE) and nonresidential fixed investment—as core drivers of employment. Reid integrates traditional macroeconomic theories, notably Keynesian principles, emphasizing that consumption and investment are principal determinants of employment levels. The study's strength lies in its historical background via integration of Keynesian frameworks.

Using wavelet analysis, Mutascu and Sokic (2021) reexamine Okun's Law, which historically describes a negative correlation between unemployment and GDP. Their analysis, from 1948Q1 to 2020Q4, highlights significant frequency-dependent relationships, showing distinct short-term, medium-term, and long-term dynamics. Short-term relationships are volatile and event-driven, medium-term relationships reflect Keynesian demand-side influences, and long-term dynamics reflect classical supply-side influences. The authors' wavelet methodology provides deeper insights than traditional static regression models and highlights nonlinearities and structural sensitivities in economic relationships.

Smith, Timmermann, and Wright (2024) focus on structural shifts in the Phillips Curve using Bayesian panel regression techniques on both U.S. and EU disaggregated data (sectoral and regional levels). They find significant flattening trends around the early 2000s, notably linked to globalization and import competition from China. Their Bayesian framework allows flexible modeling of parameter heterogeneity and time instability, precisely identifying regime shifts. This structural modeling contributes by demonstrating changing relationships under globalization.

Zhang (2023) empirically assesses U.S. unemployment, specifically during the COVID-19 pandemic, using a time-series multiple linear regression from June 2019 to February 2022. This study uniquely isolates the impact of GDP, exchange rates, and trade deficits. GDP is confirmed as negatively correlated with unemployment rates, while trade deficits and exchange rates show weaker or negligible effects. Zhang's focused methodology provides targeted insights into how shocks like pandemics shape economic relationships, highlighting GDP's central role even in crisis contexts.

Ahmed and Nasser (2023) investigate the relationship between the unemployment rate and the current account balance using a threshold vector error correction model (VECM). They identify threshold cointegration, demonstrating asymmetric adjustments—unemployment declines correlate significantly with deteriorating current account balances, contradicting prevalent political rhetoric on trade negatively impacting employment. Their use of threshold VECM captures nonlinear dynamics between unemployment and macroeconomic balances, offering insightful empirical clarity against simplistic political narratives.

Across these studies, GDP is consistently the strongest predictor of unemployment, universally demonstrating a negative relationship (Green et al., 2023; Reid, 2023; Zhang, 2023). However, inflation and interest rate effects vary notably depending on methodological approaches, sample periods, and analytical granularity (Smith et al., 2024; Ahmed & Nasser, 2023). The Phillips Curve's stability, highlighted in Smith et al. (2024), contrasts with the complexity revealed by Mutascu and Sokic (2021), who emphasize frequency-dependent and nonlinear dynamics. Methodologically, studies diverge substantially from classical regression analyses (Reid, 2023; Green et al., 2023) to advanced Bayesian and wavelet techniques (Smith et al., 2024; Mutascu & Sokic, 2021). These variations affect interpretations.

Despite extensive exploration, significant gaps persist. Most notably, few studies systematically apply penalized regression methods—such as Ridge and Lasso—to predict unemployment despite their potential to manage multicollinearity and overfitting problems common in macroeconomic datasets. The reviewed literature primarily employs traditional regression methods, potentially overlooking nuanced predictive insights from advanced machine learning or penalized regression techniques. Furthermore, the integration of machine learning methods with traditional econometric models is underrepresented, particularly in unemployment prediction using macroeconomic indicators such as GDP, CPI, and interest rates.

Given recent methodological advancements and increased computational power, leveraging penalized regression approaches could enhance predictive power. This gap motivates the present investigation's contribution: employing Ordinary Least Squares (OLS), Ridge, and Lasso regressions to robustly predict unemployment rates using real GDP, CPI, and interest rates, addressing multicollinearity, ensuring robustness, and enhancing predictive accuracy. This research aims to bridge existing gaps through methodological innovation, providing policymakers and economists with  empirical tools for economic forecasting and decision-making.

**Methodology**

Below I will delineate the mathematical techniques and statistical evaluation leveraged for analysis of U.S. unemployment rates with quarterly macroeconomic data.

*Modelling Objective*

Recall that the primary objective of this project is to predict quarterly U.S. unempployment rates using three predictors:
1. Real GDP, measured in billions of chained 2017 dollars,
2. Consumer Price Index (CPI) growth rate,
3. Federal Funds Effective Interest Rate (Interest Rate).
My hypothesis is that GDP and interest rates significantly influence unemployment, while the CPI growth rate may play a context-dependent or weaker role. We will leverage supervised regression techniques to quantify and predict these relationships.

*Mathematical Methods*

1. **Our baseline model is the ordinary least squares (OLS) multivariate regression:**

$$
\hat{y} = \beta_0 + \beta_1 \cdot \text{GDP} + \beta_2 \cdot \text{CPI} + \beta_3 \cdot \text{Interest}
$$

where:

- $\hat{y}$ is the predicted unemployment rate  
- $\beta_0$ is the intercept  
- $\beta_1, \beta_2, \beta_3$ are the estimated coefficients  

The objective is to minimize the residual sum of squares (RSS):

$$
\text{RSS} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$


This model assumes linearity, independence of errors, homoscedasticity, and no perfect multicollinearity.

2. **We will build upon our baseline model via Ridge regression.**

Ridge regression introduces L2 regularization, modifying the loss function as follows:

$$
\text{Loss}_{\text{Ridge}} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \alpha \sum_{j=1}^{p} \beta_j^2
$$

where $\alpha$ is the regularization strength. Ridge is useful when predictors are correlated, as it shrinks coefficients without eliminating them.

3. **We will lastly test Lasso regression.**

Lasso regression introduces L1 regularization, modifying the loss function as follows:

$$
\text{Loss}_{\text{Lasso}} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \alpha \sum_{j=1}^{p} |\beta_j|
$$

This encourages sparsity by shrinking some coefficients exactly to zero, which can be useful for feature selection.


Unlike Ridge, Lasso can set some coefficients to exactly zero, performing implicit feature selection.

*Implementation*
- All datasets were downloaded from the FRED database.  
- GDP, CPI, and Fed Funds rate were merged with the unemployment rate by quarter.  
- Features were standardized using `StandardScaler()` prior to Ridge and Lasso.

*Model Fitting*
- OLS was fit using both `statsmodels` (for inference) and `sklearn` (for prediction).  
- Ridge and Lasso were fit using `RidgeCV` and `LassoCV`.

*Evaluation Metrics*

For all models, the following metrics were used:

- **Coefficient of Determination** ($R^2$): Measures the proportion of variance

in unemployment explained by the model.

  $$
  R^2 = 1 - \frac{\text{SS}_{\text{res}}}{\text{SS}_{\text{tot}}}
  $$

- **Root Mean Squared Error (RMSE)**: Measures the average prediction error magnitude.

  $$
  RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
  $$

- **Mean Squared Error (MSE)** and the **Durbin–Watson statistic** were also used to detect heteroscedasticity and autocorrelation.

*Model Comparison Strategy*

Each model was evaluated on its ability to predict unemployment in the 2013-2018 period, including:

- Multivariate OLS (baseline)  
- Ridge (regularized, multicollinearity robustness)  
- Lasso (feature selection)

The GDP coefficient remained dominant, while the CPI and interest rate coefficients were further reduced in magnitude, though still retained in the model.
The Lasso regression model introduced greater sparsity by applying an L1 penalty. This led to excluding CPI growth as a predictor, effectively setting its coefficient to zero. The model retained GDP and the federal funds rate as active predictors, with GDP continuing to exhibit a strong and negatively signed coefficient. The Lasso model's R² was slightly lower at 0.981, and the RMSE increased marginally to 0.1475. Despite this slight decline in fit statistics, the Lasso model offered the advantage of greater parsimony with minimal loss in explanatory accuracy.

*Interpretation*

The empirical findings from all three models converge on a central conclusion: real GDP stands out as the most stable and economically significant predictor of unemployment from 2013 through 2018. Its consistently negative and statistically significant coefficient across all models reinforces the inverse relationship articulated by Okun's Law, which posits that increases in aggregate output are associated with decreases in unemployment due to rising labor demand.
The OLS and Ridge models each achieved an R² value of 0.982 and an RMSE of 0.1436, indicating a nearly identical and exceptionally high level of in-sample explanatory accuracy. Lasso regression, by contrast, slightly underperformed on both fit metrics—yielding an R² of 0.981 and an RMSE of 0.1475, yet introduced a more compact model specification. The subdued impact of this variable aligns with the findings of Ahmed and Nasser (2023), who document attenuated interest rate effects in the post-crisis recovery period.
These results further support the view that the CPI-unemployment relationship has weakened in recent years, consistent with the literature on the structural flattening of the Phillips Curve. The decision of the Lasso model to exclude CPI reinforces the argument advanced by Smith, Timmermann, and Wright (2024), who highlight the diminishing role of inflation expectations in labor market dynamics.
Overall, the statistical consistency of the GDP coefficient, combined with its theoretical alignment with output-employment models, substantiates its importance as a leading indicator of unemployment. The relative underperformance and diminished relevance of CPI and interest rates, particularly in the Lasso framework, suggest that simpler, output-centered models may be sufficient and preferable in similar macroeconomic contexts.

**Limitations and Future Research**

*Limitations*

Despite the strong in-sample performance of the models, several limitations should be acknowledged:
1. Small Sample Size:


*   The analysis was based on quarterly data, resulting in a relatively small number of observations. With 20 data points, statistical power is limited, and the model may be more sensitive to individual observations or outliers.


2. Macro-Level Aggregation


*   The dataset reflects national-level trends, potentially obscuring important regional, industry-specific, or demographic variation within unemployment. Labor market behavior often differs by state, sector, and age group, which the model does not consider.


3. Omitted Variable Bias


*   Several factors known to influence unemployment were not included for model simplification. This may include wage growth, labor force participation, productivity, job vacancy rates, or fiscal stimulus measures. Exclusion of these variables may introduce bias into the beta-estimates of the predictors.


4. Needing Validation


*   The models were evaluated in-sample without out-of-sample forecasting or validation. As a result, generalizability to new or future periods remains untested within the analysis.


5. Temporal Relationships


*  The models assume that GDP, interest rates, and inflation affect unemployment within the same quarter. In reality, macroeconomic policy variables, such as the interest rate, may operate with lags, meaning that the impact is distributed across time. This requires a dynamic model -- not a static one as leveraged above.


*Future Research*

To build upon the foundation established in this study, future work can pursue several avenues.
1.   Time Series or Dynamic Models


*   Incorporating lagged variables or dyanmic time series models, such as ARIMAX or vector autoregressions, would capture the delayed effects of macroeconomic shocks, better modelling the persistence of unemployment over time.


2.   Expanding Variables and Sample


*   Introducing additional macroeconomic and labor market indicators -- such as labor force participation, wage growth, or consumer sentiment -- may increase model explanatory power and reduce omitted variable bias.

3. Increase Temporal Timeframe


*   Using monthly data would increase the sample size and allow for more flexible modeling. Alternatively, using state-level data could enable panel regression techniques to uncover geographic heterogeneity in unemployment dynamics.


4. Cross-Validation Testing


*   Future studies should evaluate models on withheld data or future time perios to assess predictive accuracy or robstness. This would provide insights during economic turning points in regards to the model's explanatory power.


5. Nonlinear or Complex Methods


*   To capture complex relationships or interactions between variables, researchers may leverage nonlinear models, such as a Random Forest, Gradient Boosting, or Neural Network model. Though these models are difficult to interpret, they provide increased forecasting applications.

**Conclusion**

This investigation sought to evaluate whether real GDP, CPI growth, and the federal fund's effective interest rate could reliably predict the U.S. unemployment rate from 2013–2018. Across all models, Ordinary Least Squares (OLS), Ridge, and Lasso, real GDP consistently emerged as the most statistically and economically significant predictor. The OLS and Ridge models produced an R² of 0.982 with an RMSE of 0.1436, indicating a high level of explanatory precision. While slightly lower in fit (R² = 0.981, RMSE = 0.1475), Lasso regression generated a more parsimonious model by excluding CPI growth altogether. These results affirm the primary hypothesis that GDP, inflation, and interest rates influence unemployment trends, with GDP demonstrating the most robust effect.

This study also validates the second hypothesis: penalized regression techniques can marginally improve predictive efficiency while reducing model complexity. Ridge regression retained all predictors with reduced variance, while Lasso’s exclusion of CPI growth aligns with broader academic findings. As such, Smith, Timmermann, and Wright (2024) emphasize that the Phillips Curve has structurally flattened, rendering inflation less predictive of unemployment, a shift empirically reflected in Lasso’s results. Furthermore, the muted influence of interest rates is consistent with Ahmed and Nasser’s (2023) research on monetary policy's diminished, nonlinear effects during low-rate environments.

In addressing a methodological gap in the literature—one often limited to traditional OLS frameworks (e.g., Green et al., 2023; Zhang, 2023)—this project demonstrates that regularization methods offer theoretical and practical macroeconomic forecasting advantages. These models provide interpretable, statistically grounded tools capable of handling multicollinearity and overfitting, two common issues in economic modeling.
Future research can extend this work by incorporating higher-frequency data, testing alternative inflation measures, or integrating nonlinear modeling techniques such as elastic nets or ensemble methods. Additionally, including time-varying parameter models may reveal how economic relationships evolve under different policy regimes.
Ultimately, models that balance parsimony, rigor, and empirical accuracy hold substantial long-term value. As governments and financial institutions navigate uncertain macroeconomic conditions, such frameworks can inform real-time decision-making, optimize resource allocation, and enhance labor market policies.
