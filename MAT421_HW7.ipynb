```python
import numpy as np
import matplotlib.pyplot as plt

# Define f(x)
def f(x):
    return np.sin(x)

# forward finite difference approximation
def forward_fd(f, x, h):
    return (f(x + h) - f(x)) / h

# Define x and step size h
x = np.linspace(0, 10, 100)
h = 0.01

# Compute derivative approximation
df_approx = forward_fd(f, x, h)

# Exact derivative
df_exact = np.cos(x)

# Plot
plt.plot(x, df_approx, label="Forward FD Approximation")
plt.plot(x, df_exact, label="Exact Derivative", linestyle='dashed')
plt.legend()
plt.xlabel("x")
plt.ylabel("Derivative")
plt.title("Finite Difference Approximating First Derivative")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define f(x)
def f(x):
    return np.sin(x)

# Central difference approximation of the derivative
def central_fd(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

# Define x and step size h
x = np.linspace(0, 10, 100)
h = 0.5

# Compute derivative approximation
df_approx = central_fd(f, x, h)

# Exact derivative
df_exact = np.cos(x)

# Plot
plt.plot(x, df_approx, label="Central FD Approximation")
plt.plot(x, df_exact, label="Exact Derivative", linestyle='dashed')
plt.legend()
plt.xlabel("x")
plt.ylabel("Derivative")
plt.title("Finite Difference Approximating First Derivative (Central Difference)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define f(x)
def f(x):
    return np.sin(x)

# Central difference approximation of second derivative
def second_derivative_fd(f, x, h):
    return (f(x + h) - 2 * f(x) + f(x - h)) / (h**2)

# Define x and h
x = np.linspace(0, 10, 100)
h = 0.9

# Compute second derivative
d2f_approx = second_derivative_fd(f, x, h)

# Exact second derivative
d2f_exact = -np.sin(x)

# Plot
plt.plot(x, d2f_approx, label="Second Derivative Approximation")
plt.plot(x, d2f_exact, label="Exact Second Derivative", linestyle='dashed')
plt.legend()
plt.xlabel("x")
plt.ylabel("Second Derivative")
plt.title("Finite Difference Approximating Second Derivative")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define f(x)
def f(x):
    return np.sin(x)

# Higher-order central difference for first derivative
def high_order_central_fd(f, x, h):
    return (-f(x + 2*h) + 8*f(x + h) - 8*f(x - h) + f(x - 2*h)) / (12 * h)

# Define x and h
x = np.linspace(0, 10, 100)
h = 1.5

# Compute the derivative approximation
df_approx = high_order_central_fd(f, x, h)

# Exact derivative
df_exact = np.cos(x)

# Plot
plt.plot(x, df_approx, label="High-Order Central FD Approximation")
plt.plot(x, df_exact, label="Exact Derivative", linestyle='dashed')
plt.legend()
plt.xlabel("x")
plt.ylabel("Derivative")
plt.title("Higher-Order Central Difference for First Derivative")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define f(x)
def f(x):
    return np.sin(x)

# Finite difference approximation for the third derivative
def third_derivative_fd(f, x, h):
    return (f(x + 2*h) - 3*f(x + h) + 3*f(x - h) - f(x - 2*h)) / (2 * h**3)

# Define x and h
x = np.linspace(0, 10, 100)
h = 0.6

# Compute the third derivative approximation
d3f_approx = third_derivative_fd(f, x, h)

# Exact third derivative
d3f_exact = -np.cos(x)

# Plot
plt.plot(x, d3f_approx, label="Third Derivative Approximation")
plt.plot(x, d3f_exact, label="Exact Third Derivative", linestyle='dashed')
plt.legend()
plt.xlabel("x")
plt.ylabel("Third Derivative")
plt.title("Finite Difference Approximating Third Derivative")
plt.show()
```
